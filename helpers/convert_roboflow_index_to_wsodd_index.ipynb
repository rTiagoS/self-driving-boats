```python
def get_wsodd_index(x):
    
    if x == '0':
        return '2'
    elif x == '1':
        return '13'
    elif x == '2':
        return '7'
    elif x == '3':
        return '4'
    elif x == '4':
        return '1'

    else:
        print("erro")
        return None
    
```

```python
index = 0

labels = []

with open('/usr/src/dataset/tcc_dataset/val/labels_bkp/2022-07-17_21-53_png.rf.db54cdf6000fdfd946f3cc28a053e4ac.txt') as f:
    lines = f.readlines()
    
for line in lines:
    roboflow_class = line[0]
    wsodd_class = get_wsodd_index(roboflow_class)
    line = line[:index] + f'{wsodd_class}' + line[index+1:]
    labels.append(line)

labels
```

```python
# open file in write mode
with open('/usr/src/dataset/tcc_dataset/val/labels/2022-07-17_21-53_png.rf.db54cdf6000fdfd946f3cc28a053e4ac.txt', 'w') as fp:
    for line in labels:
        # write each item on a new line
        fp.write("%s\n" % line)
    print('Done')
```

```python
index = 0
output_dir_name = '/usr/src/dataset/tcc_dataset/val/labels'
count = 0
for dirname, _, filenames in os.walk(path):
    for filename in filenames:
        count += 1
        labels = []
        # open file in read mode
        with open(os.path.join(dirname, filename)) as f:
            lines = f.readlines()
    
            for line in lines:
                roboflow_class = line[0]
                wsodd_class = get_wsodd_index(roboflow_class)
                line = line[:index] + f'{wsodd_class}' + line[index+1:]
                labels.append(line)
            
        # open file in write mode
        with open(os.path.join(output_dir_name, filename), 'w') as fp:
            for line in labels:
                # write each item on a new line
                fp.write("%s\n" % line)
            print(f'Done: {count}')
```
